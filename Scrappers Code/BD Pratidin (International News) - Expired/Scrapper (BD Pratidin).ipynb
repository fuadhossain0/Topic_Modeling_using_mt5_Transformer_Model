```python
import requests
from bs4 import BeautifulSoup
import json
from IPython.display import clear_output
```

```python
path = "D:\\Thesis\\Scrapper\\"
news_file = path + "news_data.json"
iteration_file = path + "iteration_data.txt"
```

```python
news_list = []
with open(news_file, "r", encoding="utf-8") as fp:
  news_list = json.load(fp)
print(len(news_list))
```

```python
iter_data = 0
with open(iteration_file, "r", encoding="utf-8") as fpit:
  iter = json.load(fpit)
print(iter)
```

```python
print(len(news_list))
print(iter)
print(iter_data)
```

```python
ii = iter
cnt = 0
limit = 1000
while 1:
  if(cnt == limit):
    break
  news_list_temp = []
  page_link = 'https://www.bd-pratidin.com/international-news/' + str(ii)
  page = requests.get(page_link)
  soup = BeautifulSoup(page.content, 'html.parser')
  link_part_arr = soup.select("div.col-md-3 > a")
  main_page_link = 'https://www.bd-pratidin.com/'
  for f in link_part_arr:
    link = main_page_link + f.get('href')
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    headline = soup.select("div.area-curve > h1")[0].get_text().strip()
    news = soup.select("article.mt-3 > p")
    full_news = ""
    for x in range(0, len(news)-1):
      full_news = full_news + " " + news[x].get_text().strip()
    full_news = full_news.replace("&nbsp;", " ")
    full_news = full_news.replace("\xa0", " ")
    full_news = full_news.replace("  ", " ")
    full_news = full_news.replace("   ", " ")
    full_news = full_news.replace("  ", " ")
    dic = {"title": headline, "content": full_news}
    news_list_temp.append(dic)
  ii = ii + 14
  clear_output(wait=False)
  print(cnt)
  news_list = news_list + news_list_temp
  iter_data = ii
  cnt += 1
```

```python
# Write iteration value into iteration file
with open(iteration_file, "w", encoding="utf-8") as fpit:
  json.dump(iter_data, fpit, ensure_ascii=False)
# Write news_list into news file
with open(news_file, "w", encoding="utf-8") as fp:
  json.dump(news_list, fp, ensure_ascii=False)
```
